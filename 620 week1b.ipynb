```python
import matplotlib.pyplot as plt
import networkx as nx

# initialising the Graph
G = nx.krackhardt_kite_graph()

g = nx.generators.krackhardt_kite_graph()



person = {2: "Carol", 5:"Fernando", 7: "Heather", 8:"Ike", 9:"Jane", 6:"Garth", 4:"Ed", 1:"Beverly", 3:"Diane", 0:"Andre"}

# Creating the Graph Object

G1 = nx.Graph()

# create edge and assign manualy
# Andre connect to Beverly, Carol, Diane, Fernando
# Berverly connect to Diane, Ed, Garth
# Carol connect to Ed, Fernando, Garth
# Ed connect to Garth
# Fernando connect to Garth, Heather
# Garth connect to Heather
# Heather connect to Ike
# Ike connect to Jane

edge = [(0,1),(0,2),(0,3),(0,5),(1,3),(1,4),(1,6),(2,3),(2,5),(3,4),(3,5),(3,6),(4,6),(5,6),(5,7),(6,7),(7,8),(8,9)]

G1.add_edges_from(edge)

# assign person to the nodes

H=nx.relabel_nodes(G1,person)

# draw the graph

nx.draw(H, with_labels=1)
plt.show()
```
